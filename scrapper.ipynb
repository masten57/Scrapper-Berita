### Request

```python
#send request to the url
req=requests.get(url)
soup = BeautifulSoup(req.text, 'html.parser')  # atau 'html5lib'
soup
```

```python
# Temukan elemen div utama yang memiliki class 'row col-offset-fluid clearfix js-giant-wp-sticky-parent'
container = soup.find('div', class_='row col-offset-fluid clearfix js-giant-wp-sticky-parent')

# Debugging: Pastikan elemen ditemukan
if container:
    print("Elemen utama ditemukan!")
else:
    print("Elemen utama TIDAK ditemukan!")

# Ambil semua link artikel dalam container ini
link_page = []
if container:
    # Cari semua tag <a> dengan atribut href di dalam elemen ini
    links = container.find_all('a', href=True)
    
    print(f"\nDitemukan {len(links)} link dalam container.")  # Debug jumlah link
    
    for link in links:
        href = link['href']
        print(f"Link ditemukan: {href}")  # Debug untuk melihat setiap link
        
        # Menambahkan link ke dalam list jika belum ada (hindari duplikasi)
        if href not in link_page:
            link_page.append(href)

# Cetak semua link artikel yang berhasil dikumpulkan
print("\nDaftar Link Artikel:")
if link_page:
    for link in link_page:
        print(link)
else:
    print("Tidak ada link ditemukan.")
```
